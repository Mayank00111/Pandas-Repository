```python
#total number of people in each country who worked with Programming Language

country_respondent = df['Country'].value_counts()
```

```python
country_respondent
```

```python
country_uses_Python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())   
#total number of people in eaach country who worked with Python
```

```python
country_uses_Python
```

```python
python_df = pd.concat([country_respondent,country_uses_Python], axis = 'columns', sort = False)
```

```python
python_df   #Combine DataFrame objects with overlapping columns and return everything. Columns outside the intersection will be filled with NaN values.
```

```python
python_df.rename(columns={'Country': 'NumRespondent', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace = True)
```

```python
python_df
```

```python
python_df['PercentKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondent']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by = 'PercentKnowsPython', ascending = False)
```
